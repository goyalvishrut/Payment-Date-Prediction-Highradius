```python
# importing important library used

import pandas as pd
import numpy as np
import datetime
from sklearn import preprocessing
from sklearn.model_selection import train_test_split
import math
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.linear_model import LinearRegression
from sklearn.svm import SVR
from sklearn.tree import DecisionTreeRegressor
from sklearn.ensemble import RandomForestRegressor
from sklearn.metrics import mean_squared_error
from sklearn.metrics import r2_score
from sklearn.metrics import accuracy_score
from sklearn.preprocessing import StandardScaler
from sklearn.feature_selection import SelectFromModel
from sklearn.feature_selection import SelectKBest
from sklearn.feature_selection import chi2
import xgboost as xgb
from scipy import stats
```

```python
# reading the dataset provided

df=pd.read_csv("1805663.csv")
df.head()
```

```python
#converting to date format

df["due_in_date"]=pd.to_datetime(df["due_in_date"],format = "%Y%m%d")
df["document_create_date.1"]=pd.to_datetime(df["document_create_date.1"],format = "%Y%m%d")
df["clear_date"]=pd.to_datetime(df["clear_date"])
df["posting_date"]=pd.to_datetime(df["posting_date"])
df["baseline_create_date"]=pd.to_datetime(df["baseline_create_date"],format = "%Y%m%d")

print(df.shape)
df.head()
```

```python
#splitting the dataset into train data and test data based on clear date null and non null value

test_set=df.loc[df["clear_date"].isnull()].reset_index(drop=True)
prev_data=df.loc[df["clear_date"].notnull()].reset_index(drop=True)
```

```python
#finding the delay column (target column)

prev_data["delay"]=(prev_data["clear_date"]-prev_data["due_in_date"]).dt.days
prev_data.head()
```

```python
# sorting the data based on document create date basis

prev_data.sort_values(by="document_create_date.1",inplace=True)
prev_data.head()
```

# Spliting 

```python
# spliting the data into train set, val1, val2

d1=prev_data['document_create_date.1'].quantile(0.70)
d2=prev_data['document_create_date.1'].quantile(0.85)
d3=prev_data['document_create_date.1'].quantile(1)

train_set=prev_data.loc[prev_data['document_create_date.1']<=d1].reset_index(drop=True)
val1=prev_data.loc[(prev_data['document_create_date.1']>d1)&(prev_data['document_create_date.1']<=d2)].reset_index(drop=True)
val2=prev_data.loc[(prev_data['document_create_date.1']>d2)&(prev_data['document_create_date.1']<=d3)].reset_index(drop=True)
```

# Feature Engineering

```python
# extracting month, quarter, week from due date

train_set['due_date_quarter']=train_set['due_in_date'].dt.quarter
train_set['due_date_month']=train_set['due_in_date'].dt.month
train_set['due_date_week']=train_set['due_in_date'].dt.week
train_set.head()
```

```python
train_set['invoice_currency'].unique()
```

```python
train_set['business_code'].value_counts()
```

```python
# encoding the business code and invoive currency

coder={"business_code":     {"U001":0, "U002":1, "U005":1, "U007":1, "U013":2, "CA02":3}, "invoice_currency": {"USD": 0, "CAD": 1}}
train_set = train_set.replace(coder)
```

```python
# target encoding business code with delay column(target)

delay_mapper_business_code = train_set.groupby(['business_code'])['delay'].mean().to_dict()
delay_mapper_business_code
```

```python
# target encoding custmer payment terms with delay column(target)

delay_mapper_terms = train_set.groupby('cust_payment_terms')['delay'].mean().to_dict()
delay_mapper_terms
```

```python
# target encoding custmer number with delay column(target)

delay_mapper_number = train_set.groupby('cust_number')['delay'].mean().to_dict()
delay_mapper_number
```

```python
train_set.head()
```

```python
train_set["business_code"].value_counts()
```

```python
# extracting difference between posting,baseline and due date

train_set['due_posting']=(train_set['due_in_date']-train_set['posting_date']).dt.days
train_set['due_baseline']=(train_set['due_in_date']-train_set['baseline_create_date']).dt.days
```

```python
train_set['total_open_amount'].describe()
```

```python
train_set['due_posting'].describe()
```

```python
# mapping the features created above

train_set["business_code_delay"] = train_set['business_code'].map(delay_mapper_business_code)
train_set["terms_delay"] = train_set['cust_payment_terms'].map(delay_mapper_terms)
train_set["number_delay"] = train_set['cust_number'].map(delay_mapper_number)
train_set.describe()
```

# Feature Selection

```python
# making a dataframe for feature selection convenience

data=train_set.loc[:,['business_code','invoice_currency','total_open_amount','delay','due_date_month','due_date_week','due_posting','due_baseline','business_code_delay','terms_delay','number_delay']]
```

```python
# feature selection for linear regression model

model = LinearRegression()

# fit the model to start training.
model.fit((data.drop(['delay'],axis=1)),train_set.loc[:,'delay'])

# get the importance of the resulting features.
importances = model.coef_

# create a data frame for visualization.
final_df = pd.DataFrame({"Features": (data.drop(['delay'],axis=1)).columns, "Importances":importances})
final_df.set_index('Importances')

# sort in descending order 
final_df = final_df.sort_values('Importances',ascending=False)

#visualising feature importance
pd.Series(model.coef_, index=(data.drop(['delay'],axis=1)).columns).nlargest(5).plot(kind='barh')  
final_df
```

```python
# feature selection for decision ttree regression model

model = DecisionTreeRegressor()
model.fit(data.drop(['delay'],axis=1),train_set.loc[:,'delay'])
print(model.feature_importances_)
feat_importances = pd.Series(model.feature_importances_, index=(data.drop(['delay'],axis=1).columns))
feat_importances.nlargest(10).plot(kind='barh')
plt.show()
```

```python
# making a corelation matrix

corr=train_set.corr()
plt.figure(figsize=(10,10))
sns.heatmap(corr,annot=True)
```

```python
# making x_train, y_train for model training

x_train=train_set.loc[:,['due_posting','business_code_delay','terms_delay','number_delay']]
y_train=train_set.loc[:,['delay']]
x_train.head()
```

## Mapping the Val1 value 

```python
val1['due_posting']=(val1['due_in_date']-val1['posting_date']).dt.days
val1 = val1.replace(coder)
```

```python
val1["business_code_delay"] = val1['business_code'].map(delay_mapper_business_code)
val1["terms_delay"] = val1['cust_payment_terms'].map(delay_mapper_terms)
val1["number_delay"] = val1['cust_number'].map(delay_mapper_number)
val1.isnull().sum()
```

```python
# null value handling in val1 features set

val1['terms_delay'].fillna(train_set['terms_delay'].mode()[0],inplace=True)
val1['number_delay'].fillna((val1['number_delay'].mean()),inplace=True)
val1.isnull().sum()
```

```python
# making x_val1, y_val1

x_val1=val1.loc[:,['due_posting','business_code_delay','terms_delay','number_delay']]
y_val1=val1.loc[:,'delay']
```

## Mapping the Val2 value

```python
val2['due_posting']=(val2['due_in_date']-val2['posting_date']).dt.days
val2 = val2.replace(coder)
```

```python
val2["business_code_delay"] = val2['business_code'].map(delay_mapper_business_code)
val2["terms_delay"] = val2['cust_payment_terms'].map(delay_mapper_terms)
val2["number_delay"] = val2['cust_number'].map(delay_mapper_number)
val2.isnull().sum()
```

```python
# null value handling in val2 features set

val2['terms_delay'].fillna(train_set['terms_delay'].mode()[0],inplace=True)
val2['number_delay'].fillna((val2['number_delay'].mean()),inplace=True)
val2.isnull().sum()
```

```python
# making x_val2, y_val2

x_val2=val2.loc[:,['due_posting','business_code_delay','terms_delay','number_delay']]
y_val2=val2.loc[:,'delay']
```

# Implementation of different models

```python
MSE_Score1 = []
R2_Score1 = []
Algorithm = []
MSE_Score2 = []
R2_Score2 = []
RMSE1 = []
RMSE2 = []
```

# Linear Regression

```python
# Fitting Linear Regression to the Training Set

Algorithm.append('Linear Regression')
clf = LinearRegression()
clf.fit(x_train, y_train)

# Predicting the Test Set Results

predicted1 = clf.predict(x_val1)
predicted2 = clf.predict(x_val2)
predicted3 = clf.predict(x_train)

MSE_Score1.append(mean_squared_error(y_val1, predicted1))
R2_Score1.append(r2_score(y_val1, predicted1))
RMSE1.append(math.sqrt(mean_squared_error(y_val1, predicted1)))

MSE_Score2.append(mean_squared_error(y_val2, predicted2))
R2_Score2.append(r2_score(y_val2, predicted2))
RMSE2.append(math.sqrt(mean_squared_error(y_val2, predicted2)))
```

# Decision Tree Regressor

```python
# Fitting Decision Tress Regressor to the Training Set

Algorithm.append('Decision Tree Regressor')
clf = DecisionTreeRegressor()
clf.fit(x_train, y_train)

# Predicting the Test Set Results

predicted1 = clf.predict(x_val1)
predicted2 = clf.predict(x_val2)
predicted3 = clf.predict(x_train)

MSE_Score1.append(mean_squared_error(y_val1, predicted1))
R2_Score1.append(r2_score(y_val1, predicted1))
RMSE1.append(math.sqrt(mean_squared_error(y_val1, predicted1)))

MSE_Score2.append(mean_squared_error(y_val2, predicted2))
R2_Score2.append(r2_score(y_val2, predicted2))
RMSE2.append(math.sqrt(mean_squared_error(y_val2, predicted2)))
```

# XGB Regressor

```python
# Fitting XGB Regressor to the Training Set

Algorithm.append('XGB Regressor')
clf = xgb.XGBRegressor()
clf.fit(x_train, y_train)

# Predicting the val1 and val2 Results

predicted1 = clf.predict(x_val1)
predicted2 = clf.predict(x_val2)

MSE_Score1.append(mean_squared_error(y_val1, predicted1))
R2_Score1.append(r2_score(y_val1, predicted1))
RMSE1.append(math.sqrt(mean_squared_error(y_val1, predicted1)))

MSE_Score2.append(mean_squared_error(y_val2, predicted2))
R2_Score2.append(r2_score(y_val2, predicted2))
RMSE2.append(math.sqrt(mean_squared_error(y_val2, predicted2)))
```

# Random Forest Regressor

```python
# Fitting Random Forest Regressor to the Training Set

Algorithm.append('Random Forest Regressor')
clf = RandomForestRegressor()
clf.fit(x_train, y_train.values.ravel())

# Predicting the val1 and val2 Results

predicted1 = clf.predict(x_val1)
predicted2 = clf.predict(x_val2)

MSE_Score1.append(mean_squared_error(y_val1, predicted1))
R2_Score1.append(r2_score(y_val1, predicted1))
RMSE1.append(math.sqrt(mean_squared_error(y_val1, predicted1)))

MSE_Score2.append(mean_squared_error(y_val2, predicted2))
R2_Score2.append(r2_score(y_val2, predicted2))
RMSE2.append(math.sqrt(mean_squared_error(y_val2, predicted2)))
```

# Support Vector Regression

```python
# Fitting SVR to the Training Set

Algorithm.append('Support Vector Regression')
clf = SVR()
clf.fit(x_train, y_train.values.ravel())

# Predicting the val1 and val2 Results

predicted1 = clf.predict(x_val1)
predicted2 = clf.predict(x_val2)

MSE_Score1.append(mean_squared_error(y_val1, predicted1))
R2_Score1.append(r2_score(y_val1, predicted1))
RMSE1.append(math.sqrt(mean_squared_error(y_val1, predicted1)))

MSE_Score2.append(mean_squared_error(y_val2, predicted2))
R2_Score2.append(r2_score(y_val2, predicted2))
RMSE2.append(math.sqrt(mean_squared_error(y_val2, predicted2)))
```
